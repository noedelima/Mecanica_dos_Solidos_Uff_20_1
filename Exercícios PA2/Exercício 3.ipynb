# Viga $ABC$

A partir das equações de equilíbrio, temos:

\begin{equation*}
    \sum_{M_{z}} = 0 \therefore
    V_{B}\times 2,4m - 70kN\times 4,8m = 0
\end{equation*}

Logo,

\begin{equation*}
    V_{B} = 140kN
\end{equation*}

E,

\begin{equation*}
    \sum_{F_{y}} = 0 \therefore
    V_{C} + V_{B} - 70kN = 0
\end{equation*}

Portanto,

\begin{equation*}
    V_{C} = -70kN
\end{equation*}

Orientando o eixo $x$ com origem em $C$, no sentido de $A$, temos:

\begin{equation*}
    V\left(x\right) = V_{C} + V_{B}\left<x-2,4\right>^{0}
\end{equation*}

\begin{equation*}
    M\left(x\right) = \frac{d^{2}y\left(x\right)}{dx^{2}} = V_{C}\cdot x + V_{B}\left<x-2,4\right>
\end{equation*}

Neste caso não há constante de integração por não haver momento aplicado (ou reações de momento).

\begin{equation*}
    EI\cdot \theta\left(x\right) = \frac{dy\left(x\right)}{dx} = \frac{V_{C}}{2}\cdot x^{2} + \frac{V_{B}}{2}\left<x-2,4\right>^{2} + C_{1}
\end{equation*}

\begin{equation*}
    EI\cdot y\left(x\right) = \frac{V_{C}}{6}\cdot x^{3} + \frac{V_{B}}{6}\left<x-2,4\right>^{3} + C_{1}\cdot x + C_{2}
\end{equation*}

As condições de contorno nesta viga são:

* $EI\cdot y\left(0\right) = EI\cdot y_{C}$

* $EI\cdot y\left(2,4\right) = EI\cdot y_{B}$

Portanto, $C_{2} = EI\cdot y_{C}$, e:

\begin{equation*}
    EI\cdot y\left(2,4\right) = \frac{V_{C}}{6}\cdot \left(2,4\right)^{3} + \frac{V_{B}}{6}\left<2,4-2,4\right>^{3} + C_{1}\cdot 2,4 + EI\cdot y_{C} = EI\cdot y_{B}
\end{equation*}

```python
EI,x,V_B,V_C,C,y_B,y_C = sp.symbols("EI,x,V_B,V_C,C,y_B,y_C")
y = (1/EI)*((V_C/6)*sp.SingularityFunction(x,0,3) + (V_B/6)*sp.SingularityFunction(x,2.4,3) + C*x + EI*y_C)
display(sp.Eq(EI*y.subs(x,2.4),EI*y_B))
display("C_1 =")
display(sp.solve(sp.Eq(EI*y.subs(x,2.4),EI*y_B),C)[0])
```

# Viga $DBE$

Pelas equações de equilíbrio, bem como por simetria,

\begin{equation*}
    V_{D} = V_{E} = \frac{V_{B}}{2} = 70kN
\end{equation*}

Portanto, orientando o eixo $x$ com origem em $D$, no sentido de $E$, temos:

\begin{equation*}
    V\left(x\right) = V_{D} - V_{B}\left<x-3\right>^{0}
\end{equation*}

\begin{equation*}
    M\left(x\right) = \frac{d^{2}y\left(x\right)}{dx^{2}} = V_{D}\cdot x - V_{B}\left<x-3\right>
\end{equation*}

Neste caso também não há constante de integração por não haver momento aplicado (ou reações de momento).

\begin{equation*}
    EI\cdot \theta\left(x\right) = \frac{dy\left(x\right)}{dx} = \frac{V_{D}}{2}\cdot x^{2} - \frac{V_{B}}{2}\left<x-3\right>^{2} + C_{1}
\end{equation*}

\begin{equation*}
    EI\cdot y\left(x\right) = \frac{V_{D}}{6}\cdot x^{3} - \frac{V_{B}}{6}\left<x-3\right>^{3} + C_{1}\cdot x + C_{2}
\end{equation*}

As condições de contorno nesta viga são:

* $EI\cdot y\left(0\right) = 0$

* $EI\cdot \theta\left(3\right) = 0$

* $EI\cdot y\left(6\right) = 0$

Portanto, $C_{2} = 0$

\begin{equation*}
    EI\cdot \theta\left(3\right) = \frac{70}{2}\cdot 3^{2} + C_{1} = 0 \therefore
\end{equation*}

\begin{equation*}
     C_{1} = -315
\end{equation*}

Temos, então, que:

\begin{equation*}
    EI\cdot y_{B} = EI\cdot y\left(3\right) = \frac{V_{D}}{6}\cdot 3^{3} -315\cdot 3 = \frac{70\times 3^{3}}{6} - 315\times 3 = - 630
\end{equation*}

# Viga $FCG$

A viga $FCG$ é semelhante à viga $DBE$, com a diferença da carga $V_{C}=-70kN$ em lugar de $V_{B}$. Portanto,

\begin{equation*}
    V_{F} = V_{G} = \frac{V_{C}}{2} = -35kN
\end{equation*}

Portanto, orientando o eixo $x$ com origem em $F$, no sentido de $G$, temos:

\begin{equation*}
    V\left(x\right) = V_{F} - V_{C}\left<x-3\right>^{0}
\end{equation*}

\begin{equation*}
    M\left(x\right) = \frac{d^{2}y\left(x\right)}{dx^{2}} = V_{F}\cdot x - V_{C}\left<x-3\right>
\end{equation*}

Neste caso também não há constante de integração por não haver momento aplicado (ou reações de momento).

\begin{equation*}
    EI\cdot \theta\left(x\right) = \frac{dy\left(x\right)}{dx} = \frac{V_{F}}{2}\cdot x^{2} - \frac{V_{C}}{2}\left<x-3\right>^{2} + C_{1}
\end{equation*}

\begin{equation*}
    EI\cdot y\left(x\right) = \frac{V_{F}}{6}\cdot x^{3} - \frac{V_{C}}{6}\left<x-3\right>^{3} + C_{1}\cdot x + C_{2}
\end{equation*}

As condições de contorno nesta viga são:

* $EI\cdot y\left(0\right) = 0$

* $EI\cdot \theta\left(3\right) = 0$

* $EI\cdot y\left(6\right) = 0$

Portanto, $C_{2} = 0$

\begin{equation*}
    EI\cdot \theta\left(3\right) = -\frac{35}{2}\cdot 3^{2} + C_{1} = 0 \therefore
\end{equation*}

\begin{equation*}
     C_{1} = -\frac{315}{2}
\end{equation*}

Temos, então, que:

\begin{equation*}
    EI\cdot y_{C} = EI\cdot y\left(3\right) = \frac{V_{F}}{6}\cdot 3^{3} = -\frac{35\times 3^{3}}{6} + \frac{315\times 3}{2} = 315
\end{equation*}
