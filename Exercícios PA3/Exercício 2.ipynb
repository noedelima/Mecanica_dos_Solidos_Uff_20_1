# Introdução

Temos, para cada tipo de esforço, as seguintes equações de energia (para consulta).

## Esforço Normal

\begin{equation*}
    U_{N} = \int_{0}^{L}{\frac{N}{2EA}}dx =
    \frac{N^{2}L}{2EA}
\end{equation*}

## Esforço Cortante

\begin{equation*}
    U_{V} = \int_{0}^{L}{\frac{V^{2}x^{2}}{2EI}}dx =
    \frac{V^{2}L^{3}}{6EI}
\end{equation*}

## Momento Fletor

\begin{equation*}
    U_{M} = \int_{0}^{L}{\frac{M^{2}}{2EI}}dx =
    \frac{M^{2}L}{2EI}
\end{equation*}

## Momento Torsor

\begin{equation*}
    U_{T} = \int_{0}^{L}{\frac{T^{2}}{2GJ}}dx =
    \frac{T^{2}L}{2GJ}
\end{equation*}
