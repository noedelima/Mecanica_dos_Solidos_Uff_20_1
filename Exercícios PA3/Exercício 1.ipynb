# Introdução

Temos, para cada tipo de esforço, as seguintes equações de energia (para consulta).

## Esforço Normal

\begin{equation*}
    U_{N} = \int_{0}^{L}{\frac{N}{2EA}}dx =
    \frac{N^{2}L}{2EA}
\end{equation*}

## Esforço Cortante

\begin{equation*}
    U_{V} = \int_{0}^{L}{\frac{V^{2}x^{2}}{2EI}}dx =
    \frac{V^{2}L^{3}}{6EI}
\end{equation*}

## Momento Fletor

\begin{equation*}
    U_{M} = \int_{0}^{L}{\frac{M^{2}}{2EI}}dx =
    \frac{M^{2}L}{2EI}
\end{equation*}

## Momento Torsor

\begin{equation*}
    U_{T} = \int_{0}^{L}{\frac{T^{2}}{2GJ}}dx =
    \frac{T^{2}L}{2GJ}
\end{equation*}

a) Sistema Principal

Para resolver este sistema, vamos desconsiderar $C$ como um apoio, e inserir em seu lugar duas cargas que são, respectivamente, as reações de apoio $V_{C}(\uparrow)$ e $H_{C}(\rightarrow)$.

b) Condições de Compatibilidade de Deformação

Da mesma forma, vamos considerar as duas restrições de movimento $\delta_{Cx}=\delta_{Cy}=0$.

c) Equações do Teorema de Castigliano

\begin{equation*}
    \delta_{Cx} = \frac{\partial U}{\partial H_{C}} = 0
\end{equation*}

\begin{equation*}
    \delta_{Cy} = \frac{\partial U}{\partial V_{C}} = 0
\end{equation*}

As equações de equilíbrio fornecem:

\begin{equation*}
    \sum_{F_{x}} = 0 \therefore
    H_{A} + H_{C} = -P \therefore
    H_{A} = -P - H_{C}
\end{equation*}

\begin{equation*}
    \sum_{F_{y}} = 0 \therefore
    V_{A} + V_{C} = 0 \therefore
    V_{A} = -V_{C}
\end{equation*}

\begin{equation*}
    \sum_{M_{z}} = 0 \therefore
     -LH_{C} + LV_{C} + M_{A} = \frac{PL}{2} \therefore
     M_{A} = \frac{PL}{2} + LH_{C} - LV_{C}
\end{equation*}

Onde temos cinco variáveis ($H_{A}$, $V_{A}$, $M_{A}$, $H_{B}$, $V_{B}$), e apenas três equações. Para resolver o sistema, necessitamos das duas equações dadas pelo teorema de Castigliano citadas acima.

### Equação de Energia
Utilizando apenas a energia de momento fletor (desprezando-se a energia relativa ao esforço cortante), temos:

\begin{equation*}
    U_{1} = \int_{0}^{\frac{L}{2}}\left[\frac{\left(H_{A}x\right)^{2}}{2EI} + \frac{M_{A}^{2}}{2EI}\right]dx =
    \frac{1}{2EI}\int_{0}^{\frac{L}{2}}\left[H_{A}^{2}x^{2} + M_{A}^{2}\right]dx =
    \frac{1}{2EI}\left[\frac{H_{A}^{2}}{3}x^{3} + M_{A}^{2}x\right]_{0}^{\frac{L}{2}} =
    \frac{1}{2EI}\left[\frac{H_{A}^{2}}{24}L^{3} + \frac{M_{A}^{2}}{2}L\right]
\end{equation*}

\begin{equation*}
    U_{2} = \int_{0}^{\frac{L}{2}}\left[\frac{\left(H_{C}x\right)^{2}}{2EI} + \frac{\left(V_{C}L\right)^{2}}{2EI}\right]dx =
    \frac{1}{2EI}\left[\frac{H_{C}^{2}}{3}x^{3} + V_{C}^{2}L^{2}x\right]_{0}^{\frac{L}{2}} =
    \frac{1}{2EI}\left[\frac{H_{C}^{2}}{24}L^{3} + \frac{V_{C}^{2}}{2}L^{3}\right]
\end{equation*}

\begin{equation*}
    U_{3} = \int_{0}^{\frac{L}{2}}\frac{\left(V_{C}x\right)^{2}}{2EI}dx =
    \frac{1}{2EI}\left[\frac{V_{C}^{2}}{3}x^{3}\right]_{0}^{\frac{L}{2}} =
    \frac{1}{2EI}\left[\frac{V_{C}^{2}}{24}L^{3}\right]
\end{equation*}

Logo,

\begin{equation*}
    U = \sum_{i=1}^{3}{U_{i}} =
    \frac{1}{2EI}\left[\frac{H_{A}^{2}}{24}L^{3} + \frac{M_{A}^{2}}{2}L +
    \frac{H_{C}^{2}}{24}L^{3} + \frac{V_{C}^{2}}{2}L^{3} +
    \frac{V_{C}^{2}}{24}L^{3}\right] \therefore
\end{equation*}

\begin{equation*}
    U = 
    \frac{1}{2EI}\left[\frac{\left(-P-H_{C}\right)^{2}}{24}L^{3} + \frac{\left(\frac{PL}{2}+LH_{C}-LV_{C}\right)^{2}}{2}L +
    \frac{H_{C}^{2}}{24}L^{3} + \frac{V_{C}^{2}}{2}L^{3} +
    \frac{V_{C}^{2}}{24}L^{3}\right] \therefore
\end{equation*}

\begin{equation*}
    U =
    \frac{L^{3}}{48EI}\left[\left(-P-H_{C}\right)^{2} + 12\left(\frac{P}{2}+H_{C}-V_{C}\right)^{2} +
    H_{C}^{2} + 13V_{C}^{2}\right]
\end{equation*}

Para resolver o sistemas, basta derivar $U$ em relação a $H_{C}$ e $V_{C}$ e igualar ambas as expressões a zero para obter as equações pendentes que tornam o sistema solucionável.

```python
L,EI,P = sp.symbols('L,EI,P', positive=True)
H_C,V_C = sp.symbols('H_C,V_C')
U = ((L**3)/(48*EI))*((-P-H_C)**2 + 12*(P/2+H_C-V_C)**2 + H_C**2 + 13*V_C**2)
display(U,sp.diff(U,H_C),sp.diff(U,V_C))
```

Logo, Temos,

\begin{equation*}
    \frac{L^{3}}{24EI}\left(14H_{C}+7P-12V_{C}\right) = 0 \therefore
\end{equation*}

\begin{equation*}
    14H_{C}+7P-12V_{C} = 0
\end{equation*}

E

\begin{equation*}
    \frac{L^{3}}{24EI}\left(-12H_{C}-6P+25V_{C}\right) = 0 \therefore
\end{equation*}

\begin{equation*}
    -12H_{C}-6P+25V_{C} = 0
\end{equation*}

Este conjunto de equações permite resolver $H_{C}$ e $V_{C}$ que, por sua vez, permitem resolver $H_{A}$, $V_{A}$ e $M_{A}$.

```python
K = sp.Matrix([[14,-12],[-12,25]])
f = sp.Matrix([[-7*P],[6*P]])
K.solve(f)
```

Portanto, $H_{C}=-\frac{P}{2}$ e $V_{C}=0$. Logo, $H_{A}=H_{C}=-\frac{P}{2}$, $V_{A}=V_{C}=0$ e $M_{A}=0$.

\begin{equation*}
    \theta_{B} = \frac{\partial U}{\partial M_{B}} = KM_{B}
\end{equation*}

### Equação de Energia
Utilizando apenas a energia de momento fletor (desprezando-se a energia relativa ao esforço cortante), temos:

\begin{equation*}
    U = \int_{0}^{L}\left[\frac{\left(V_{A}x\right)^{2}}{2EI} + \frac{M_{A}^{2}}{2EI} + \frac{\left(\frac{qx^{2}}{2}\right)^{2}}{2EI}\right]dx =
    \frac{1}{2EI}\left[\frac{V_{A}^{2}}{3}x^{3} + M_{A}^{2}x + \frac{q^{2}}{6}x^{3}\right]_{0}^{L} =
    \frac{L}{2EI}\left[\frac{V_{A}^{2}}{3}L^{2} + M_{A}^{2} + \frac{q^{2}}{6}L^{2}\right]
\end{equation*}
